# Profile Averaging Plot
In this notebook, we show multivariate ClaSP profiles and their average.

```python
# needed for PDF/A
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', None)

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
sns.set_color_codes()

from src.utils import load_raw_has_datasets
from claspy.clasp import ClaSP

import os
import sys

sys.path.append("../../")
```

Let's start by loading the data and selecting a time series.

```python
df_data = load_raw_has_datasets()
df_data.head()
```

```python
idx = 0
row = df_data.iloc[idx, :]

T = row.iloc[8:17].tolist()
```

We compute the profiles and plot them.

```python
P = []

for time_series in T:
    profile = ClaSP(window_size=50).fit_transform(time_series)
    P.append(profile)
    
p_avg = np.mean(np.array(P), axis=0)
```

```python
title = f"ClaSP Profiles for TS#{row.ts_id+1} ({', '.join(row.activities)})"
time_series = P + [np.zeros(0), np.zeros(0),p_avg]
sensor_names = [""] * (len(time_series)-1) + ["AVG"] # df_data.columns[8:17].tolist()
change_points = row.change_points
activities = row.activities
font_size = 18

fig, axes = plt.subplots(
    len(time_series),
    sharex=True,
    gridspec_kw={'hspace': .1},
    figsize=(20, len(time_series) // 2)
)

activity_colours = {}
idx = 0

for activity in activities:
    if activity not in activity_colours:
        activity_colours[activity] = f"C{idx}"
        idx += 1

for idx, (ts, sensor, ax) in enumerate(zip(time_series, sensor_names, axes)):
    if len(ts) > 0:
        segments = [0] + change_points.tolist() + [ts.shape[0]]
        for kdx in np.arange(0, len(segments) - 1):
            ax.plot(
                np.arange(segments[kdx], segments[kdx + 1]),
                ts[segments[kdx]:segments[kdx + 1]],
                c=activity_colours[activities[kdx]],
                alpha=0.5 if idx < len(time_series) - 3 else 1
            )
        
        ax.axvline(x=np.argmax(ts), linewidth=2, color='red', linestyle="dashed")

    ax.set_ylabel(sensor, fontsize=12)

    for tick in ax.xaxis.get_major_ticks():
        tick.label1.set_fontsize(font_size)

    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_fontsize(font_size)
        
    ax.set_facecolor((1., 1., 1.))
        
    ax.tick_params(
        axis='both',          
        which='both',      
        labelbottom=False,
        labelleft=False
    )

axes[0].set_title(title, fontsize=18)

plt.savefig("../figures/profile_averaging.pdf", bbox_inches="tight")
```
